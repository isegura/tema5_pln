
## 2.Carga del tokenizador:
Utiliza la clase BertTokenizer de la librería transformers para cargar el tokenizador del modelo bert-base-uncased.


```python
from transformers import BertTokenizer

# Cargar el tokenizador de BERT para español
tokenizer_spanish = BertTokenizer.from_pretrained('dccuchile/bert-base-spanish-wwm-cased')

print("Tokenizador BERT para español cargado correctamente.")
```

## 3. Tokenización

Se proporcionarán una serie de oraciones de ejemplo.
Para cada oración, utiliza el tokenizador para obtener la lista de tokens.
Imprime la oración original y la lista de tokens resultantes.

Ejemplos de oraciones:

- "El perro corre rápido."
- "¡Hola, mundo!"
- "Esta es una palabra desconocida: xyzzy."
- "La palabra compuesta es: anti-inflamatorio."


```python
# Oraciones de ejemplo
oraciones_ejemplo = [
    "El perro corre rápido.",
    "¡Hola, mundo!",
    "Esta es una palabra desconocida: xyzzy.",
    "La palabra compuesta es: anti-inflamatorio."
]

# Tokenizar las oraciones y mostrar los resultados
for oracion in oraciones_ejemplo:
    tokens = tokenizer_spanish.tokenize(oracion)
    print(f"Oración original: {oracion}")
    print(f"Tokens: {tokens}")
    print("-" * 30)
```

## 5. Experimentación

Experimenta con diferentes oraciones, incluyendo oraciones con palabras desconocidas, caracteres especiales y diferentes estructuras gramaticales.

Observa cómo cambia la tokenización con diferentes oraciones.

Intenta tokenizar oraciones en diferentes idiomas (si utilizas un modelo BERT multilingüe).



```python
# Más oraciones para experimentar
oraciones_experimentacion = [
    "El auto rojo está aparcado en la calle principal.",
    "¿Qué tal si vamos al café 'El Rincón' mañana a las 3:30 p.m.?",
    "Ella dijo: '¡Qué día tan bonito!' y sonrió.",
    "Mi computadora portátil tiene 16GB de RAM y un procesador i7.",
    "La palabra inventada 'supercalifragilístico' es muy larga.",
    "#Colab es una herramienta genial para aprender ML."
]

print("--- Experimentación con nuevas oraciones ---")
for oracion in oraciones_experimentacion:
    tokens = tokenizer_spanish.tokenize(oracion)
    print(f"Oración original: {oracion}")
    print(f"Tokens: {tokens}")
    print("-" * 30)

print("--- Observaciones ---")
print("1. Las palabras desconocidas o compuestas pueden ser divididas en subtokens (sub-word units), indicadas por '##'.")
print("2. Los signos de puntuación y caracteres especiales suelen ser tokenizados como unidades separadas.")
print("3. El tokenizador `dccuchile/bert-base-spanish-wwm-cased` es 'cased', lo que significa que distingue entre mayúsculas y minúsculas (por ejemplo, 'El' y 'el' serían tokens diferentes en teoría, aunque en este modelo puede que las trate de forma similar si son parte de palabras comunes).")
print("4. Para idiomas diferentes al español, este tokenizador específico no funcionará de manera óptima; se necesitaría un tokenizador multilingüe o específico para ese idioma.")
```
