# Ejemplo 3 — Prompting para *sentiment analysis*

Este notebook ilustra **prompting**:
usar un modelo ya entrenado/instruction-tuned **sin volver a entrenarlo**.

## Objetivo
Resolver el mismo problema de análisis de sentimiento usando solo prompts.

## Idea clave
- **No cambiamos el modelo**
- **No hay entrenamiento adicional**
- Solo cambiamos la **entrada**

## Qué diferencia este notebook
Frente a los otros dos:

- **Fine-tuning clásico** → reentrena para una tarea
- **Instruction fine-tuning** → reentrena para seguir instrucciones
- **Prompting** → no reentrena nada; solo formula bien la petición

```
# Si hace falta, descomenta:
# !pip install -q transformers sentencepiece

from transformers import pipeline
```

```
model_name = "google/flan-t5-small"

generator = pipeline(
    "text2text-generation",
    model=model_name,
    tokenizer=model_name,
)
```

## Prompt simple

```
reviews = [
    "I loved the movie, it was emotional and beautifully acted.",
    "The film was boring and far too long.",
    "The visuals were impressive but the story was weak.",
]

for review in reviews:
    prompt = f"Classify the sentiment of this review as positive or negative: {review}"
    result = generator(prompt, max_new_tokens=5)[0]["generated_text"]
    print(f"Review: {review}\nPrediction: {result}\n")
```

## Comparación: prompt pobre vs prompt mejorado

Una parte importante del prompting es la **formulación**.

```
review = "The acting was good, but the script was dull and repetitive."

bad_prompt = f"Sentiment: {review}"
good_prompt = (
    "Classify the sentiment of the following movie review. "
    "Answer with exactly one word: positive or negative. "
    f"Review: {review}"
)

bad_result = generator(bad_prompt, max_new_tokens=10)[0]["generated_text"]
good_result = generator(good_prompt, max_new_tokens=5)[0]["generated_text"]

print("BAD PROMPT")
print(bad_prompt)
print("Output:", bad_result)
print()
print("GOOD PROMPT")
print(good_prompt)
print("Output:", good_result)
```

## Prompting con pocos ejemplos (*few-shot prompting*)

También podemos dar ejemplos dentro del propio prompt.

```
few_shot_prompt = '''
Classify the sentiment of each review as positive or negative.

Review: I loved every minute of this film.
Answer: positive

Review: The movie was awful and painfully slow.
Answer: negative

Review: The soundtrack was excellent but the plot was disappointing.
Answer:
'''.strip()

result = generator(few_shot_prompt, max_new_tokens=5)[0]["generated_text"]
print(result)
```

## Conclusión

Este notebook representa **prompting**:

- usamos un modelo ya disponible;
- **no** hacemos entrenamiento adicional;
- el rendimiento depende en parte de **cómo formulamos el prompt**.

Es la forma más directa de ilustrar que en prompting **no cambiamos el modelo, cambiamos la entrada**.
